### Figure 3.2 - Regression coefficients - RSS
Note that the text in the book describes the coefficients based on uncentered data, whereas the plot shows the model based on centered data. The latter is visually more appealing for explaining the concept of a minimum RSS. I think that, in order not to confuse the reader, the values on the axis of the B0 coefficients have been changed to correspond with the text. The axes on the plots below are unaltered.

```python
# Regression coefficients (Ordinary Least Squares)
regr = skl_lm.LinearRegression()

# X = scale(advertising.TV, with_mean=True, with_std=False).reshape(-1,1)
X = scale(advertising.TV, with_mean=False, with_std=False).reshape(-1,1)
# X = advertising.TV.values.reshape(-1, 1)
y = advertising.Sales

regr.fit(X,y)
print(regr.intercept_)
print(regr.coef_)
```

```python
# Create a grid of coordinates for plotting
B0 = np.linspace(regr.intercept_-2, regr.intercept_+2, 50)
B1 = np.linspace(regr.coef_-0.02, regr.coef_+0.02, 50)
xx, yy = np.meshgrid(B0, B1, indexing='xy')
Z = np.zeros((B0.size, B1.size))

# Calculate Z-values (RSS) based on grid of coefficients
n = X.shape[0]
for (i, j), v in np.ndenumerate(Z):
    y_hat = xx[i, j] + yy[i, j]*X.ravel()
    Z[i, j] = ((y - y_hat)**2).sum()
    
# Minimized RSS
min_RSS = r'$\beta_0$, $\beta_1$ for minimized RSS'
y_hat = regr.intercept_ + regr.coef_* X.ravel()
min_rss = np.sum((y - y_hat)**2)
min_rss
```

```python
fig = plt.figure(figsize=(15, 6))
fig.suptitle('RSS - Regression coefficients', fontsize=20)

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='3d')

# Left plot
CS = ax1.contour(xx, yy, Z, cmap=plt.cm.Set1)
ax1.scatter(regr.intercept_, regr.coef_[0], c='r', label=min_RSS)
ax1.clabel(CS, inline=True, fontsize=10, fmt='%1.1f')

# Right plot
ax2.plot_surface(xx, yy, Z, alpha=0.3)
ax2.contour(xx, yy, Z, zdir='z', offset=Z.min(), cmap=plt.cm.Set1, alpha=0.4)
ax2.scatter3D(regr.intercept_, regr.coef_[0], zs=min_rss, c='r', label=min_RSS)
ax2.set_zlabel('RSS')
ax2.set_zlim(Z.min(), Z.max())

# setting common to both the plots
for ax in fig.axes:
    ax.set_xlabel(r'$\beta_0$', fontsize=17)
    ax.set_ylabel(r'$\beta_1$', fontsize=17)
    ax.legend()
    
plt.show()
```

+ How accurate sample mean $\hat{\mu}$ is estimate of population mean $\mu$? We know that the average of $\hat{\mu}$'s over many data sets will be very close to $\mu$, but that a single estimate $\hat{\mu}$ may be a substantial over-estimate or under-estimate of $\mu$. How far off that single estimate of $\hat{\mu}$ be? In general, we answer this question by computing the *standard error* of $\hat{\mu}$, written as $SE(\hat{\mu})$. We have the well-known formula 

$$ 
Var(\hat{\mu}) = SE(\hat{\mu})^2 = \frac{\sigma^2}{n}
$$

<u><b>Proof</b></u>: 

$$
\begin{align}
Var(\hat{\mu}) &= Var(\frac{x_1 + x_2 + \cdots + x_n}{n}) \\
&= Var(\frac{1}{n}x_1 + \frac{1}{n}x_2 + \cdots + \frac{1}{n}x_n) \\
&= \frac{1}{n^2}Var(x_1) + \frac{1}{n^2}Var(x_2) + \cdots + \frac{1}{n^2}Var(x_n) 
\end{align}
$$
Add $x_i$ are identically distributed, which means they have same variance. Therefore, replacing $Var(x_i)$ with the alternative notation $\sigma^2$. We get:

$$
\begin{align}
Var(\hat{\mu}) &= \frac{1}{n^2}[\sigma^2 + \sigma^2 + \cdots + \sigma^2] \\
&= \frac{\sigma^2}{n}
\end{align}
$$

The standard error tells us the average amount that this estimate $\hat{\mu}$ differs from the actual value, $\mu$ 

+ For the linear regression, $\hat{y} = \hat{\beta_0} + \hat{\beta_1}x_1 + \epsilon \text{ where, } \epsilon \sim iid \text{ } N(0, \sigma^2) $
+ Using OLS we estimate $\beta_0$ and $\beta_1$ as 
$$
\beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^{2}}\\
\beta_0 = \bar{y} - \hat{\beta_1}\bar{x}
$$

+ In similar way, we can calculate how close $\hat{\beta_0}$ and $\hat{\beta_1}$ are to the true values, $\beta_0$ and $\beta_1$. To compute the standard errors associated with these values, we use following formulas:

$$
\text{SE}(\hat{\beta_0})^2 = \sigma^2\bigg[\frac{1}{n} + \frac{\bar{x}^2}{\sum_{i=1}^{n}(x_{i} - \bar{x})^2}\bigg] \\
\text{SE}(\hat{\beta_1})^2 = \frac{\sigma^2}{\sum_{i=1}^{n}(x_{i} - \bar{x})^2} \\
$$
In general $\sigma^2$ is not known, but can be estimated from the data.  
$$
\text{where}, \sigma \text{ is known as residual standard error and is given by formula, } RSE = \sqrt{\frac{RSS}{n-2}}
$$
$95\%$ confidence interval for $\beta_1$ approximately takes the form, $\hat{\beta_1} \pm 2. SE(\hat{\beta_1})$

<u><b>Proof</b></u>: 

a.
$$
\begin{align}
\sum_{i}(x_i - \bar{x})\bar{y} &= \bar{y}\sum_{i} (x_i - \bar{x}) \\
&= \bar{y} \big(\big(\sum_{i} x_i\big) - n\bar{x}\big) \\
&= \bar{y}(n\bar{x} - n\bar{x}) \\
&= 0
\end{align}
$$
b.
$$
\begin{align}
\sum_{i}(x_i - \bar{x})(y_i - \bar{y}) &= \sum_{i}(x_i - \bar{x})y_i - \sum_{i}(x_i - \bar{x})\bar{y} \\
&= \sum_{i}(x_i - \bar{x})y_i - 0 \\
&= \sum_{i}(x_i - \bar{x})(\beta_0 + \beta_1 x_i + \epsilon_{i})
\end{align}
$$

c.
$$
\begin{align}
Var(\hat{\beta_1}) &= Var\bigg(\dfrac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n}(x_i - \bar{x})^{2}}\bigg) \\
&= Var\bigg(\dfrac{\sum_{i=1}^{n} (x_i - \bar{x})(\beta_0 + \beta_1 x_i + \epsilon_i)}{\sum_{i=1}^{n}(x_i - \bar{x})^{2}}\bigg) \text{ , substituting (b.)} \\
&= Var\bigg(\dfrac{\sum_{i=1}^{n} (x_i - \bar{x})\epsilon_i}{\sum_{i=1}^{n}(x_i - \bar{x})^{2}} \bigg) \text{ , noting only } \epsilon_i \text{ is a random variable.} \\
&= \dfrac{\sum_{i=1}^{n} (x_i - \bar{x})^2 Var(\epsilon_i)}{\big(\sum_{i=1}^{n}(x_i - \bar{x})^{2}\big)^2}, \text{independence of } \epsilon_i \text{ and, } Var(kX) = k^2 Var(X) \\
&= \frac{\sigma^2}{\sum_i (x_i - \bar{x})^2}
\end{align}
$$

d.
$$
\begin{align}
Var(\hat{\beta_0}) = Var(\bar{y} - \beta_1 \bar{x}) \\
\end{align}
$$

```python
# no. of observations
n = X.shape[0]
# Residual Standard Error, RSE
sigma = np.sqrt(min_rss/(n-2))
x_mean = np.mean(X)
# SE(b_0)**2
se_b0_2 = sigma**2 * (1/n + (x_mean**2)/np.sum((X - x_mean)**2))
# SE(b_1)**2
se_b1_2 = sigma**2 / np.sum((X - x_mean)**2)
# SE(b_0)
se_b0 = np.sqrt(se_b0_2)
# SE(b_1)
se_b1 = np.sqrt(se_b1_2)
np.round(se_b0, 3), np.round(se_b1, 3)
```

```python
# confidence interval of b_0 and b_1
(est.params[0] - 2*se_b0, est.params[0] + 2*se_b0), (est.params[1] - 2*se_b1, est.params[1] + 2*se_b1)
```

```python
# RSS with regression coefficients
((advertising.Sales - (est.params[0] + est.params[1]*advertising.TV))**2).sum()
```

$$
H_o: \text{There is no relationship between TV advertising and Sales} \\
H_a: \text{There is some relationship between TV advertising and Sales}
$$

Mathematically,
$$
H_o: \beta_1 =  0  \\
H_a: \beta_1 \neq 0 
$$

To test the null hypothesis, we need to determine whether $\hat{\beta_1}$, our estimate for $\beta_1$ is sufficiently far from zero that we can be confident that $\beta_1$ is non-zero. How far is far enough? In practice we compute a *t-statistic*, given by
$$
t = \frac{\hat{\beta_1} - 0}{SE(\hat{\beta_1})}
$$

Which measures the number of standard deviations that $\hat{\beta_1}$ is away from 0. The t-distribution has a bell shape and for values greater than 30 it is quite similar to normal distribution.    

```python
# t for b_1
t_b1 = est.params[1]/se_b1
t_b1
```

So, the probability that we will observe a test statistic $\leq$ -17.668 and $\geq$ 17.668. $P(t \leq -17.668) + P(t \geq 17.668) = 2 \times P(t \geq 17.668) = 2 \times (1 - P(t \leq 17.668))$ 

```python
# 2*(1 - P(t <= 17.668))
p = 2*(1 - stats.t.cdf(t_b1, 198))
p
```

```python
# P(T <= t_b1)
stats.t.cdf(t_b1, 198)
```

p-values is less than 0.05. So, we reject the null-hypothesis.

The coefficients for $\hat{\beta_0}$ and $\hat{\beta_1}$ are very large relative to their standard errors, so the t-statistics are also large; the probabilities of seeing such values if $H_o$ is true are virtually zero.

```python
# Create grid coordinates for plotting
B_Age = np.linspace(regr1.coef_[0]-3, regr1.coef_[0]+3, 100)
B_Limit = np.linspace(regr1.coef_[1]-0.02, regr1.coef_[1]+0.02, 100)

B_Rating = np.linspace(regr2.coef_[0]-3, regr2.coef_[0]+3, 100)
B_Limit2 = np.linspace(regr2.coef_[1]-0.2, regr2.coef_[1]+0.2, 100)

X1, Y1 = np.meshgrid(B_Limit, B_Age, indexing='xy')
X2, Y2 = np.meshgrid(B_Limit2, B_Rating, indexing='xy')
Z1 = np.zeros((B_Age.size,B_Limit.size))
Z2 = np.zeros((B_Rating.size,B_Limit2.size))

Limit_scaled = scale(credit.Limit.astype('float'), with_std=False)
Age_scaled = scale(credit.Age.astype('float'), with_std=False)
Rating_scaled = scale(credit.Rating.astype('float'), with_std=False)

# Calculate Z-values (RSS) based on grid of coefficients
for (i,j),v in np.ndenumerate(Z1):
    Z1[i,j] =((y - (regr1.intercept_ + X1[i,j]*Limit_scaled +
                    Y1[i,j]*Age_scaled))**2).sum()/1000000
    
for (i,j),v in np.ndenumerate(Z2):
    Z2[i,j] =((y - (regr2.intercept_ + X2[i,j]*Limit_scaled +
                    Y2[i,j]*Rating_scaled))**2).sum()/1000000
```
